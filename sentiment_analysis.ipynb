```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk
nltk.download('vader_lexicon')
```

```python
sid = SentimentIntensityAnalyzer()
```

```python
text ="this is a great movie!"
```

```python
scores = sid.polarity_scores(text)
```

```python
print (scores)
```

```python
comments=pd.DataFrame({
    'comment':["Chawa is the great movie with a historical story"]
})
```

```python
comments['sentiment_score'] = comments['comment'].apply(lambda x: sid.polarity_scores(x)['compound'])
```

```python
comments['sentiment']= comments['sentiment_score'].apply(lambda x: 'positive' if x>0 else ('negative' if x<0 else 'neutral'))
```

```python
print(comments)
```

```python
print(comments['sentiment_score'])
```

```python
scores = sid.polarity_scores(comments['comment'][0])
```
